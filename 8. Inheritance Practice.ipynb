# Single level 

```python
class Student(Person):
    def __init__(self,name,age,roll,cgpa):
        Person.__init__(self,name,age)
        self.rollno = roll
        self.percentage = cgpa
        
    # This display will override display() of parent class
    def display(self):
        Person.display(self)
        print('Roll number :' , self.rollno)
        print('CGPA :' , self.percentage)
```

# Multi level
1. note : If an object of a parent class is used to invoke the method, then the version in the parent class will be executed, but if an object of the subclass is used to invoke the method, then the version in the child class will be executed. In other words, it is the type of the object being referred to (not the type of the reference variable) that determines which version of an overridden method will be executed.

```python
s1 = Student('Hrithik',21,160240107062,8.3)
s1.display()
```

```python
class Subject(Student):
    def __init__(self,name,age,roll,cgpa,subject):
        Student.__init__(self,name,age,roll,cgpa)
        self.subject = subject
    # Overrides
    def display(self):
        Student.display(self)
        print('Subject :' , self.subject)

    
```

```python
sub1 = Subject('Rutul',21,160240107068,8.9,'Python')
sub1.display()
```

```python
# Object determines which method is executed i.e. parent's display() or child's display()
student2 = Student('Harsh',22,160240107051,9)
student2.display()
```

# Multiple 

### Using Protected variables and methods

```python
class Sleeping:
    def __init__(self,hrs):
        self._hrs = hrs
    def _sleep(self):
        print("This animal sleeps", self._hrs , "a day")

class Eating:
    def __init__(self,food):
        self._food = food
    def _types(self):
        print("This animal eats", self._food, "types of food")

class Animal(Sleeping,Eating):
    def __init__(self,name,hrs,food):
        self.name = name
        Sleeping.__init__(self,hrs)
        Eating.__init__(self,food)
    def display(self):
        print("Name :", self.name)
        Sleeping._sleep(self)
        Eating._types(self)
```

```python
cow = Animal("cow",6,2)
cow.display()
```

### Without using Protected variables and methods i.e. Public

```python
class Sleeping:
    def __init__(self,hrs):
        self.hrs = hrs
    def sleep(self):
        print("This animal sleeps", self.hrs , "a day")

class Eating:
    def __init__(self,food):
        self.food = food
    def types(self):
        print("This animal eats", self.food, "types of food")

# Animal does not inherit Sleeping and Eating, even then it can call variables and methods of sleeping and eating because the variableand methods are public
class Animal:
    def __init__(self,name,hrs,food):
        self.name = name
        Sleeping.__init__(self,hrs)
        Eating.__init__(self,food)
    def display(self):
        print("Name :", self.name)
        Sleeping.sleep(self)
        Eating.types(self)
```

```python
a = Animal('cow',5,3)
a.display()
```

# Hierchical Inheritance

```python
class Vehicle:
    def __init__(self):
        print("Vehicle info :")
        

class Car(Vehicle):
    def __init__(self,price):
        Vehicle.__init__(self)
        self.price = price
    #Vehicle.display(self)
    def display(self):
        print("The car has 4 tyres and price is", self.price)

class Bike(Vehicle):
    def __init__(self,price):
        Vehicle.__init__(self)
        self.price = price
    #Vehicle.display(self)
    def display(self):
        print("The bike has 2 tyres and price is", self.price)

class Rikshaw(Vehicle):
    def __init__(self,price):
        Vehicle.__init__(self)
        self.price = price
    #Vehicle.display(self)
    def display(self):
        print("The rikshaw has 3 tyres and price is", self.price)
        
a = Car(30000)
a.display()
b = Bike(10000)
b.display()
c = Rikshaw(15000)
c.display()
```

```python
class Person:
    def __init__(self,name,age,mobile):
        self.n=name
        self.a=age
        self.m=mobile
    temp=5
    def display(self):
        print("Name is",self.n,". Age is",self.a,". Mobile number is",self.m,".")
```
